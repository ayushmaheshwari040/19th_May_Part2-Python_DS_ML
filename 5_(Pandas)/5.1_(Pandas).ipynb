# iloc & loc
   iloc:
            int based location
            
          indexing:  df.iloc[row_index, col_index]
            
          slicing:   df.iloc[row_slicing, col_slicing]

    loc:
            label indexes

          indexing:  df.loc[row_index, col_index]
            
          slicing:   df.loc[row_slicing, col_slicing]
