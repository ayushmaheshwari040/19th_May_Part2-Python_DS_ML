```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix = pd.read_csv('netflix-rotten-tomatoes-metacritic-imdb.csv')
netflix
```

```python
def func(x):
    if type(x) == float:
        return False
    else:
        if 'comedy' in x.lower().split(', '):
            return True
        else:
            return False
```

```python
netflix['Genre'].apply(func)
```

```python
netflix[netflix['Genre'].apply(func)]
```
