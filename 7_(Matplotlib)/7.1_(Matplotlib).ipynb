```python
jaipur = pd.DataFrame({
    'date':pd.date_range("05/12/25", periods=7),
    'student':np.random.randint(30, 50, 7)
})
jaipur
```

```python
delhi = pd.DataFrame({
    'date':pd.date_range("05/12/25", periods=7),
    'student':np.random.randint(30, 60, 7)
})
delhi
```

```python
pune = pd.DataFrame({
    'date':pd.date_range("05/12/25", periods=7),
    'student':np.random.randint(20, 45, 7)
})
pune
```

```python
jaipur
```

```python
plt.figure(figsize=(15, 6), dpi=200)
plt.plot(jaipur['date'], jaipur['student'])
plt.show()
```

```python
plt.figure(figsize=(15, 8), dpi=200)
plt.plot(jaipur['date'], jaipur['student'], label='Jaipur')
plt.plot(delhi['date'], delhi['student'], label='Delhi')
plt.plot(pune['date'], pune['student'], label='Pune')
plt.legend()
plt.xlabel('Dates', size=15)
plt.ylabel('Count', size=15)
plt.show()
```

```python
plt.figure(figsize=(15, 8), dpi=200)
plt.plot(jaipur['date'], jaipur['student'], 'r--8', label='Jaipur')
plt.plot(delhi['date'], delhi['student'], label='Delhi')
plt.plot(pune['date'], pune['student'], label='Pune')
plt.legend()
plt.xlabel('Dates', size=15)
plt.ylabel('Count', size=15)
plt.show()
```

```python
plt.figure(figsize=(15, 8), dpi=200)
plt.plot(jaipur['date'], jaipur['student'], 'r--*', label='Jaipur')
plt.plot(delhi['date'], delhi['student'], 'b-.h',  label='Delhi')
plt.plot(pune['date'], pune['student'], 'g:H', label='Pune')
plt.legend()
plt.xlabel('Dates', size=15)
plt.ylabel('Count', size=15)
plt.show()
```

```python
fig = plt.figure(figsize=(15, 8), dpi=200)
ax = fig.gca()
ax.set_facecolor('k')
plt.plot(jaipur['date'], jaipur['student'], 'r--*', mec='y', mfc='w', ms=9, label='Jaipur')
plt.plot(delhi['date'], delhi['student'], 'b-.h', mec='y', mfc='w', ms=9, label='Delhi')
plt.plot(pune['date'], pune['student'], 'g:H', mec='y', mfc='w', ms=9, lw=3, label='Pune')
plt.legend()
plt.grid()
plt.xlabel('Dates', size=15)
plt.ylabel('Count', size=15)
plt.show()
```

```python
fig = plt.figure(figsize=(15, 8), dpi=200)
ax = fig.gca()
plt.plot(jaipur['date'], jaipur['student'], 'r--*', mec='y', mfc='w', ms=9, label='Jaipur')
plt.plot(delhi['date'], delhi['student'], 'b-.h', mec='y', mfc='w', ms=9, label='Delhi')
plt.plot(pune['date'], pune['student'], 'g:H', mec='y', mfc='w', ms=9, lw=3, label='Pune')
plt.legend()
plt.grid()
plt.xlabel('Dates', size=15)
plt.ylabel('Count', size=15)
ticks = plt.xticks()
plt.show()
```
